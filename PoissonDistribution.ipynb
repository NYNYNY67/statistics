```python
import math
import matplotlib.pyplot as plt
```

```python
def poisson_distribution(x, lam):
  return math.e**(-lam) * lam**x / math.factorial(x)
```

```python
lam = 10
x = 5
```

```python
# 一定期間に平均lam回起こる事象がx回起こる確率
poisson_distribution(x, lam)
```

```python
p_list = []
for i in range(0, 21):
  p_list.append(poisson_distribution(i, lam))
```

```python
plt.plot(p_list)
```

```python
sum = 0
for p in p_list:
  sum += p
print(sum)
```
